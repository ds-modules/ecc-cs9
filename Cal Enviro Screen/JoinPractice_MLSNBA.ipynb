```python
import pandas as pd
```

```python
nba_teams = pd.DataFrame({
    'city': ['Los Angeles', 'San Francisco', 'Portland', 'Seattle', 'San Diego'],
    'nba_team': ['Lakers', 'Warriors', 'Trail Blazers', '', '']
})
nba_teams
```

```python
mls_teams = pd.DataFrame({
    'city': ['Los Angeles', 'Portland', 'Seattle', 'San Diego'],
    'mls_team': ['LA Galaxy', 'Timbers', 'Sounders', 'Loyal SC']
})
mls_teams
```

```python
nfl_teams = pd.DataFrame({
    'city': ['Los Angeles', 'San Francisco', 'Seattle', 'Los Angeles'],
    'nfl_team': ['Rams', '49ers', 'Seahawks', 'Chargers']
})
nfl_teams
```

```python
inner_join_nba_mls = nba_teams.merge(mls_teams, on='city', how='inner')
inner_join_nba_mls
```

```python
outer_join_nba_mls = nba_teams.merge(mls_teams, on='city', how='outer')
outer_join_nba_mls
```

```python
left_join_nba_mls = nba_teams.merge(mls_teams, on='city', how='left')
left_join_nba_mls
```

```python
right_join_nba_mls = nba_teams.merge(mls_teams, on='city', how='right')
right_join_nba_mls
```
